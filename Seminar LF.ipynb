## Forklar kort hva en DataFrame er i pandas og nevn to vanlige operasjoner man gjør på en DataFrame









### LF

En DataFrame er en todimensjonal datastruktur i pandas med rader og kolonner, lik et regneark. Vanlige operasjoner inkluderer:

- Filtrering: df[df[’A’] > 0] - gjør at vi får en ny dataframe som inneholder kun rader hvor A > 0.

```{python}
df[df['A'] > 0]
```


Hva skjer:

- Vi bruker betingelsen df[’A’] > 0 for å lage en boolsk maske som velger rader der kolonnen A har verdier større enn 0.
- Vi bruker denne masken til å indeksere DataFrame df, noe som gir oss en ny DataFrame med bare de radene som oppfyller betingelsen.
- Resultatet er en DataFrame som inneholder kun radene hvor kolonnen A har verdier større enn 0.

```{python}

df.groupby('A').mean()

```


Hva skjer:

- Vi bruker groupby-metoden på DataFrame df for å gruppere radene basert på unike verdier i kolonnen A.
- Deretter bruker vi mean()-metoden for å beregne gjennomsnittet av hver gruppe for de andre kolonnene i DataFrame.
- Resultatet er en ny DataFrame der hver rad representerer en unik verdi fra kolonnen A, og kolonnene inneholder gjennomsnittsverdiene for de tilhørende gruppene.

### Anbefalt ekstraoppgave: Beregn Totalt Sparebeløp
En person setter inn $1000$ kr på en sparekonto hver måned. Banken gir $2$% årlig rente, men renten legges ikke til før på slutten av året.

Skriv en `for`-løkke som simulerer 5 år med sparing og beregner hvor mye penger personen har etter denne perioden.


````python
månedlig_sparing = 1000 # Startverdier
rente = 0.02  # 2% årlig rente
år = 5        # 5 år med sparing
saldo = 0     # Starter med 0 kronasjer

# Løkke for å beregne saldo over flere år
for i in range(1, år + 1):
    saldo += månedlig_sparing * 12          # Legg til sparing for året
    saldo *= (1 + rente)                    # Legg til årlig rente, på slutten av året.
    print(f"Etter {i} år: {saldo:.2f} kr")  # Print resultatet, med avrundede verdier
print(f"Totalt etter {år} år: {saldo:.2f} kr")  # Print totalen etter alle årene
````


## Sympy Cobb-Douglas
$$f(K, L) = A \cdot K^{\alpha} \cdot L^{\beta}$$



lag en funksjon som definerer Cobb-Douglas funksjonen, hvor vi setter $A=1$, $\alpha = 0.2$ og $\beta = 1-\alpha$.



````python
A = 1
alpha, beta = 0.2, 0.8
def cobb_douglas(K, L, A, alpha, beta):
    return A * (K ** alpha) * (L ** beta)
````

### Finn den partiell deriverte av Cobb-Douglas funksjonen (mhp $K$). 


````python
#Definer symboler for kapital (K), arbeidskraft (L), og konstanter A, alpha, beta
A, alpha, beta, K, L = sp.symbols('A alpha beta K L')

#Finn partielle deriverte med hensyn til kapital (K)
dY_dK = sp.diff(cobb_douglas(K, L, A, alpha, beta), K)
dY_dK
````


## Vi ser på et enkelt tilbuds- og etterspørselmarked:



$$Q_e = 200 - 4p, \; \; \; \; \; \;  Q_t = 10 + 6p,$$

hvor $Q_e$ er etterspørselen og $Q_t$ er tilbud og $p$ er prisen.

1. Bruk SymPy til å finne likevektsprisen og -mengden.



````python
import sympy as sp
# Definerer symboler
qe, qt,p = sp.symbols('q_e q_t p')

# Definerer tilbud og etterspørsel
qe = 200 - 4 *p
qt = 10 + 6*p

likevekt_pris = sp.solve(sp.Eq(qe,qt),p)[0]

# Finner likevektskvantum
likevekts_kvantum = qe.subs(p, likevekt_pris)

likevekts_kvantum, likevekt_pris

````

En person ønsker å spare opp $100,000$ kr til egenkapital mot boligkjøp. Hver måned setter de inn et fast beløp på kontoen sin.
Du skal lage en funksjon som:

* Bruker en while-løkke for å beregne hvor mange måneder det tar å nå $100,000$ kr.
* Bruker en for-løkke for å skrive ut saldoen for hver måned.
* Bruker if/elif/else for å gi forskjellige meldinger underveis basert på saldoen.
* Legger alt i en funksjon, slik at vi kan kalle `sparing()` for å kjøre programmet.


````python


def sparing(sparemål, saldo, måneder, månedlig_sparing):

    # While-løkke for å beregne sparing hvor mye penger spart
    while saldo < sparemål:
        måneder += 1
        saldo += månedlig_sparing

        # If/elif/else for statusmeldinger underveis i spareprosessen :)
        if saldo >= 100000:
            print(f"Måned {måneder}: Du har spart {saldo:.2f} kr! Målet er nådd!")
        elif saldo >= 50000:
            print(f"Måned {måneder}: Du har spart {saldo:.2f} kr! Halvveis der!")
        elif saldo >= 25000:
            print(f"Måned {måneder}: Du har spart {saldo:.2f} kr! God start!")
        else:
            print(f"Måned {måneder}: Sparte {månedlig_sparing} kr, saldo: {saldo:.2f} kr")

    print(f"Gratulerer! Du har nådd sparemålet på {sparemål} kr etter {måneder} måneder!")

# Kjør funksjonenen vi har lagd
sparing(100000, 0, 0, 12000)
````
